```
#now make the function of get parallel db
def get_parallel_db(remove_index,db):
  return torch.cat((db[0:remove_index],db[remove_index+1:]))[0:6]
```

```
get_parallel_db(2,data)#we still get the same functionality
```

```
#let create a list of parallel databases
def get_parallel_dbs(db):
    parallel_db=list()
    for i in range(len(db)):
          pdb=get_parallel_db(2,db)
          parallel_db.append(pdb)
    return parallel_db
```

```
pdbs=get_parallel_dbs(data)
```

```
pdbs
```

```
def create_db_and_parallel(num_entries):
  db=torch.rand(num_entries)>0.5
  pdbs=get_parallel_dbs(db)

  return db,pdbs

```

```
db,pdbs=create_db_and_parallel(20)#database with 20 entries each
db
```

```
pdbs
```

Now we intimate that quering the database doesn't leak any private information 

```
def query(db):
  return db.sum()
```

```
db_query_result=query(db)#its means 7 true values in our database
```

```
query(pdbs[3])
```

```
#let calculate how much the query changes when we remove someone from the database
max_distance=0
for pdb in pdbs:
  pdb_query_result=query(pdb)
  db_distance=torch.abs(pdb_query_result-db_query_result)
  if (db_distance>max_distance):
    max_distance=db_distance

 
max_distance#this is basically the sensitivity of our database
```
